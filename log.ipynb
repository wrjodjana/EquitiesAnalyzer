```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sp500_data = pd.read_csv('dataset/updated_sp500_companies.csv')
```

```python
print(sp500_data.columns)
```

### Calculate SMA (Simple Moving Average)

#### 1.1 Simple Moving Average across a 10 day window for S&P 500 data

```python
# Simple Moving Average
sp500_data['SMA_10'] = sp500_data['Currentprice'].rolling(window=10).mean()

# Plot
plt.figure(figsize=(12, 5))
plt.plot(sp500_data['SMA_10'], label='SMA 10')
plt.legend()
plt.show()
```

#### 1.2 Simple Moving Average across a 50 and 200 day window for S&P 500 data

```python
sp500_data['SMA_50'] = sp500_data['Currentprice'].rolling(window=50).mean()
sp500_data['SMA_200'] = sp500_data['Currentprice'].rolling(window=200).mean()

plt.figure(figsize=(14, 7))
plt.plot(sp500_data['SMA_10'], label='SMA 10')
plt.plot(sp500_data['SMA_50'], label='SMA 50')
plt.plot(sp500_data['SMA_200'], label='SMA 200')
plt.legend()
plt.show()
```
